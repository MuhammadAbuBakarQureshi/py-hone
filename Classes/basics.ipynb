```python
class Person:

    haveBones = True

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):

        print(f"Hello, My name is {self.name} and I am {self.age} year old.")
```

```python
p1 = Person("Bakar Qureshi", 20)
p2 = Person("Abdullah", 22)

p1.greet(), p1.haveBones, p2.greet(), p2.haveBones
```
